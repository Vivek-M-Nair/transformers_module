```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
model_name="google/flan-t5-large"
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForSeq2SeqLM.from_pretrained(model_name)
def job_d():
    import pandas as pd
    df=pd.read_csv('monster_com-job_sample.csv')
    for i,text in enumerate(df["job_description"][:5]):
      yield text #The key idea behind yield is that it pauses the function and returns a value to the caller. But the important part is that the state of the function is preserved. When the function is called again (usually by the loop), it resumes execution from the point where it left off, just after the yield.


def gen_qn(job_desc,temperature=0.7, top_p=0.9):
    promp= f"generate interview question for : {job_desc}" 
    input_ids=tokenizer(promp, truncation=True, padding=True, return_tensors="pt").input_ids
    outputs=model.generate(
    input_ids,
    # max_length=100,
    do_sample=True,
    temperature=temperature,
    top_p=top_p
    )
    resp=tokenizer.decode(outputs[0],skip_special_tokens=True)
    return resp
job_list=job_d()
for i,job_desc in enumerate(job_list):
    question=gen_qn(job_desc)
    print(f"question for {job_desc} :{question}")

```

```python
def qn_generates(job_desc,temperature=0.7,top_p=0.9):
    input_text=f"Generate interview question for:{job_desc}"
    input_ids=tokenizer(input_text,padding=True,truncation=True, return_tensors="pt").input_ids
    outputs=model.generate(
    input_ids,
    max_length=100,
    do_sample=True,
    temperature=temperature,
    top_p=top_p
    )
    response= tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response
df=pd.read_csv('monster_com-job_sample.csv')
for i,job_description in enumerate(df["job_description"][1:3]):
  job_desc=job_description
  question=qn_generates(job_desc)
  # print(f"question for {i}: {question}")
  df.at[i,"response"]=question

df
```
